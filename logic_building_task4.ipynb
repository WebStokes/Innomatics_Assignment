```python
def val_recharge(amount,plan):
    valid_plans=[199,299,399,599]
    if amount <= 50:
        print("Recharge amount must be at least 50")
        return False
    
    if plan not in valid_plans:
        print("Invalid plan selected. Valid plans are: 199, 299, 399, 599")
        return False
    
    print("Recharge successful!")
    return True

while True:
    amount=int(input("Enter recharge amount: "))
    plan=int(input("Select a plan (199, 299, 399, 599): "))
    if val_recharge(amount,plan):
        break   
    else:
        print("Please try again.\n")
```

```python
def check_inventory(product):
    for product,stock in product.items():
        if stock < 15:
            print(f"{product} is low in stock. Only {stock} left.")
        else:
            print(f"{product} is in stock. {stock} available.")

inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 18
}

check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0

    for mark in marks:
        total += mark
    
    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"
    
marks = [70, 85, 90, 60, 75]
result = process_result(marks)
print("Result:", result)
```

```python
def claculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        fare += fare*0.25


    return fare

while True:
    distance = float(input("Enter distance to be traveled (in km): "))
    peak = input("Is it peak hours? (yes/no): ")

    total_fare = claculate_fare(distance, peak)

    print("Total Fare:", total_fare)

    retry = input("Do you want to calculate fare again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
def check_attendance(attendance_list):

    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day=="P":
            present_days += 1
    
    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
    
attendance = ["P", "P", "P", "P", "A", "P", "P", "A", "P", "P"]
result = check_attendance(attendance)
print("Status:", result)
```

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak"
    
    has_digit=False
    has_special=False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit=True
        elif char in special_chars:
            has_special=True
    
    if has_digit and has_special:
        return "Strong"
    else:
        return " Weak"
    
password = input("Enter a password: ")
print("Password Strength:", check_password_strength(password))
```
