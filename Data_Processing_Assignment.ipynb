```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
hig_score = max(employees.values())
top_emp=[]
for names in employees:
    if employees[names] == hig_score:
        top_emp.append(names)
print("Top performers:", top_emp)
```

```python
query = "Buy mobile phone buy phone online"
query = query.lower()
query=query.split()
frequency = {}
for word in query:
    if word in frequency:
      frequency[word] +=1
    else:
      frequency[word] = 1
for word in frequency:
   if frequency[word] > 1:
      print(word,":",frequency[word])

```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []

for index in range(len(sensor_readings)):
    value = sensor_readings[index]

    if value % 2 ==0:
        valid_readings.append((index,value))

print("Valid Sensor Readings:")
print(valid_readings)
    
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain=email.split('@')[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_emails = len(emails)
for domain in domain_count:
    percentage=(domain_count[domain] / total_emails) * 100
    print(f"{domain}: {percentage:.2f}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg_sales = sum(sales) / len(sales)
threshold = avg_sales * 1.30
for day,value in enumerate(sales,start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```
