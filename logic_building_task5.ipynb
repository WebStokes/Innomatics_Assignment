```python

def analyze_post(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like

    print(f"Total likes: {total_likes}")

    if total_likes > 1000:
        print("This post is Viral Post!")
    else:
        print("This post has a Normal Engagement.")

analyze_post([100, 200, 300, 400, 50])
analyze_post([100, 200, 300, 400])
```

```python
def check_medicine_stock(stock):
    print("Medicine Stock:", stock)

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")

check_medicine_stock(6)
check_medicine_stock(15)
```

```python
def check_rainfall(rainfall_list,required_level):
    total=0

    for rainfall in rainfall_list:
        total+=rainfall
    
    average =  total / len(rainfall_list)

    print(f"Average Rainfall: {average}")

    if average > required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

check_rainfall([70, 80, 65, 75], 70)
check_rainfall([70, 8, 60, 72], 80)
```

```python
def check_duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")



check_duplicate_accounts(["user1", "user2", "user1"])
check_duplicate_accounts(["user1", "user2", "user3"])
check_duplicate_accounts(["user1", "user2"])
```

```python
def check_appointment_eligibility(age):
    print("Patient Age:", age)

    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")

check_appointment_eligibility(21)
check_appointment_eligibility(2)
```

```python
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


filter_premium_crops([1500, 2500, 3200, 1800])
filter_premium_crops([1500, 2500, 1800])
```

```python
def check_system_health(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


check_system_health(7)
check_system_health(2)
```

```python
def check_transaction(amount):
    print("Transaction Amount:", amount)

    if amount <= 50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


check_transaction(60000)
check_transaction(40000)
```

```python
def check_attendance(attended_classes, total_classes):
    percentage = (attended_classes / total_classes) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

check_attendance(80, 100)
check_attendance(60, 100)
```

```python
def calculate_electricity_bill(units):
    bill=0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:  
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

bill, status = calculate_electricity_bill(250)
print("Total Bill:", bill)
print("Usage Level:", status)
bill, status = calculate_electricity_bill(1500)
print("Total Bill:", bill)
print("Usage Level:", status)
bill, status = calculate_electricity_bill(500)
print("Total Bill:", bill)
print("Usage Level:", status)
```
